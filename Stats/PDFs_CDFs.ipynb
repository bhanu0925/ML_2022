```python
df = sns.load_dataset('iris')
```

Histograms - multiclass analysis on numerical data, histogram does not support this

```python
sns.distplot(df['petal_length'])
```

```python
sns.FacetGrid(df,hue='species',height=6).map(sns.histplot,'petal_length').add_legend()
```

```python
sns.FacetGrid(df,hue='species',height=6).map(sns.distplot,'sepal_length').add_legend()
```
