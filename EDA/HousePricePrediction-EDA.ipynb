#### Since this is a regression problem, we have to check the following while we do multivariate Analysis

- **Normality**: several statics test , like T-test replay on normal distribution, we have to check for the normality of the data
- **Homoscedacity**: error terms shoul be same accross independent variable
- **Linearity** : there is be linear relationship between x and y
- **Absence of correlated errors** : there should be no error which exhibits - >one error is co relatted to other error. 

## Check for Normality

- This can be done with two methods
- Histograms
- Normal Probability Plot - Data distribution should closely follow the diagonal that represents the normal distribution


```python
sns.distplot(df_train['SalePrice'],fit=norm)
```

```python
res = stats.probplot(df_train['SalePrice'], plot=plt)
```

##### these  blue data points are exprected to be on red line if they are normally distributed, we can see some skeness at both the side of the line

### How do we deal with Skew and kurtosis???

 - The answer is use transformation of data, those could be,
    - Log transformation
    - Squared Term Transformation
    - Box-Cox Transformation

**For the positive skew, log transforamtion works well**

```python
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```
