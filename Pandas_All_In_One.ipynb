## Filtering Dataframe on condition | Masking using boolean series

```python
## we can also find matches held in others cities by putting the logic in a function

def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]
```

```python
get_city_match_count('Mumbai')
```
