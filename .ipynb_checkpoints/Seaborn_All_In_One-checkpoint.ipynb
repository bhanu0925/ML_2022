## Relational /Statistical plots

- used for multivarient analysis
- Scatter plot
- Line plot - used for time series
- FacetGrids - plotting graphs line by line

#### Scatter plot can be plotted for both numerical and Categorical data. But widely used for plotting numerical data.


### Two ways to draw a scatter plot in seaborn

- relplot - figure level function
- scatterplot  - axis level function
- parameters - hue, size,style

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

## Using relplot

### plot scatter between total_bill  and tips

```python
sns.relplot(x='total_bill',y='tip',data=data)
```

- relplot - has parameter 'kind' which is set to scatter by default


## The hue parameter

 - see the distrubution on the basis of smoker

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker')
```

 - red and blue point represents smoker or non smoker


### Style parameter 
 - Find the smoker is boy or girl
 
 - for style and hue , we normally pass categorical data

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker', style='sex')
```

## The Size parameter

- normal used with continuous numerical data

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker', style='sex', size = 'size')
```

## Using the scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker', style='sex', size = 'size')
```

## categorical Scatter plot

 - categorical data on x axis,
 - numerical data on y axis

```python
tip = sns.load_dataset('tips')
```

## Strip Plot

```python
sns.catplot(x='day',y='tip',kind='strip',data=tip)
```

 - jitter - parametr , induces small noise

```python
sns.catplot(x='day',y='tip',kind='strip',data=tip,jitter=10)
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=tip,jitter=0)
```

## Swarmplot

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tip)
```

- swarm plot also shows the distribution of the data in both direction x and y

### Hue parameter

```python
sns.catplot(x = 'day', y = 'tip', data =tip, hue='sex', kind='swarm')
```

```python
sns.swarmplot(x='day',y='tip',data=tip)  ## with out kind parameter
```

## Bar plot and Countplot

- these give info on central tendency

```python
sns.catplot(x = 'smoker',y ='total_bill',data=tip,kind='bar')

## black colur bars are know as error bars, showing condidence intervals to calculate mean for the population
## ci = 95, there is 95% of confidence that mean of the population lie int this reagion
```

```python
sns.catplot(x = 'smoker',y ='total_bill',hue ='sex',data=tip,kind='bar')
```

```python
sns.catplot(x = 'smoker',y ='total_bill',hue ='sex',data=tip,kind='bar',estimator=np.median)
```

```python
sns.catplot(x = 'smoker',y ='total_bill',hue ='sex',data=tip,kind='bar',estimator=np.std)
```

```python
sns.catplot(x = 'smoker',y ='total_bill',hue ='sex',data=tip,kind='bar',estimator=np.var)
```

## Count Plot

```python
sns.catplot(x ='sex',kind='count',data=tip)
```

```python
sns.catplot(x ='sex',kind='count',data=tip,hue='smoker')
```

## Heatmap

- drawn on descrete numerical data or
- drawn on two categorical data

```python
fly = sns.load_dataset('flights')
```

```python
fly
```

```python
x = fly.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,linewidths=0.7)
```

```python
sns.heatmap(x,linewidths=0.7,annot=True,fmt='d')
```

```python
sns.heatmap(x,linewidths=0.7,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,linewidths=0.7,annot=True,fmt='d',cmap='winter')
```

```python
sns.heatmap(x,linewidths=0.7,annot=True,fmt='d',cmap='summer')
```

## Cluster map

 - its heatmap , where its rows and columns are clustered with similarity

```python
sns.clustermap(x)
```

### cluster happens based on the distance bwtween two data points, by default metric parameter is eucludian distance

```python
sns.clustermap(x,metric = 'correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True,col_cluster=False,metric = 'correlation')  ## normalizing the score by using z score
```

```python
sns.clustermap(x,z_score=0,annot=True,row_cluster=False,metric = 'correlation')  ## normalizing the score by using z score
```

## Distplots

 - distplot are used to do univarient analysis on a continous/numerical data
 
 - its a combination of 3 graphs
  - Histogram
  - KDE
  - Rugplot

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.displot(titanic['age'])
```

```python
sns.displot(titanic['age'], kde=True)
```

```python
sns.displot(titanic['age'],bins=100, kde=True)
```

```python
sns.displot(titanic['age'], rug=True, kde=False)
```

```python
sns.rugplot(titanic['age'])
```

### How to use displot

```python
sns.displot(titanic[titanic['survived'] ==1]['age'],kde=True)
sns.displot(titanic[titanic['survived'] ==0]['age'],kde=True)
```

- histplot : Plot a histogram of binned counts with optional normalization or smoothing.
- kdeplot : Plot univariate or bivariate distributions using kernel density estimation.
- rugplot : Plot a tick at each observation value along the x and/or y axes.
- ecdfplot : Plot empirical cumulative distribution functions.
- jointplot : Draw a bivariate plot with univariate marginal distributions
